#### Задание
В этом задании вам необходимо составить свой дизайн A/B-теста. Для этого вам нужно выбрать произвольный продукт (например - это маркет-плейс, сервис рекомендации видео, социальная сеть, мобильное приложение или проект, над которым работаете), дать его краткое описание (основная функция продукта и его цель). Далее к этому продукту необходимо сформулировать дизайн A/B-теста, то есть определить следующие пункты:
Изменение - она же гипотеза, которая сделает продукт лучше.
Метрики, по которым будет определяться, что продукт стал лучше.
Аудитория - на ком должна быть протестирована гипотеза.
Инфраструктура - как будут доставляться изменения и как будет деление аудитории происходить. (Описать примерно)
Длительность проведения теста. (Пока не умеем ее рассчитывать, поэтому указать исходя из логических соображение)
Ожидаемый эффект - на сколько должны вырасти ключевые метрики

##### Продукт
Стриминговый сервис прослушивания музыки с рекомендациями следующего трека и возможностью ставить лайк на понравившийся трек

#### дизайн A/B-теста
##### Гипотеза:
Применениее нового алгоритма рекомендаций улучшит пользовательский опыт: пользователям будут даваться более релевантные рекомендации, тем самым увеличится время использования сервиса


##### Метрики:
1. Среднее время прослушивания рекомендовнного нового(без лайка) трека
2. Среднее количество лайков по предложенным новым трекам
3. Среднее время проведенное пользователем на сервисе


##### Аудитория:
Активные пользователи, которые пользуются сервисом больше месяца


##### Инфраструктура:
Для подходящеей аудитории будет внедрен новый алгоритм рекомендаций. Деление  подходящей аудитории будет осуществляться по остатку от деления на два хеша идентификатора: контрольную группу и экспериментальную. В зависимости от группы пользователя конфигурация сервиса будет разной: для экспериментальной группы треки будет рекомендовать новый алгоритм, для контрольной - алгоритм рекомендаций останется прежним


##### Длительность проведения теста:
6 недель


##### Ожидаемый эффект:
Прирост метрик:
1. на 20%
2. на 6%
3. на 3%
